```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
#Prediction of Customer Churn means our beloved customers withthe intention of leaving us in the future.
```

```python
data = pd.read_csv("Predict Customer Churn.csv")
data
```

```python
#removing unnecessary columns
cleanDF = data.drop('customerID', axis=1)
#Convert all the non-numeric columns to numeric
for column in cleanDF.columns:
  if cleanDF[column].dtype == np.number:
    continue
  cleanDF[column] = LabelEncoder().fit_transform(cleanDF[column])
cleanDF.dtypes
```

```python
cleanDF
```

```python
# scaling data
# Data scaling is the process of transforming the values of the features of a dataset till they are within a specific range
x = cleanDF.drop('Churn', axis=1)
y = cleanDF['Churn']
x = StandardScaler().fit_transform(x)
```

```python
# Split the data into 80% training and 20% testing

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
# Train the model
model.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)

# print the predictions
print(predictions)
```

```python
print(classification_report(ytest, predictions))
```
